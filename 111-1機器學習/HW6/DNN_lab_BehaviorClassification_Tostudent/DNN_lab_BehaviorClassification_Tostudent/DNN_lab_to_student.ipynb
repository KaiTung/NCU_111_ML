# Training phase

### Load Data

```python
# load train.csv and training-label.csv
raw_train_x = pd.read_csv("train.csv")
raw_train_y = pd.read_csv("train_label.csv")
# print(raw_train_x)
# print(raw_train_y)
```

### Data Preprocessing

```python
# drop created_time column
raw_train_x.drop('created_time',axis=1,inplace=True)
print(raw_train_x.info())
print(raw_train_y.info())
```

```python
train_x = raw_train_x.drop(['sniffer_loc'], axis=1)
train_x = train_x.groupby(['mac_hash']).sum()

for i in range(1,15):
    train_x[i] = 0

print(train_x.info())

# for i in range(33227):
#     print(str(raw_train_x.loc[i]['mac_hash']))
#     print(raw_train_x.loc[i,'sniffer_loc'])
#     if i == 20:
#         break
```

```python
# Count whether each mac_hash has passed through these 14 locations, and generate a training dataframe


for i in range(33227):
    train_x.loc[raw_train_x.loc[i]['mac_hash'],raw_train_x.loc[i,'sniffer_loc']]=1
    
print(train_x.head(15))
```

```python
# Add label column
train_x = pd.merge(train_x,raw_train_y,on='mac_hash')
train_x
```

### Training

```python
# X: feature from training dataframe
X = train_x.iloc[:,:-1]
X
```

```python
# Y: label from training dataframe
# you need to change your label from categorical encoding to one-hot encoding
Y = train_x.iloc[:,-1:]
Y
```

```python
# split your dataset as training dataset and validation dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
print(X_train)
```

```python
print(X_train.info())
```

```python
from sklearn import preprocessing
minmax_scale = preprocessing.MinMaxScaler(feature_range=(0,1))
scaledFeatures=minmax_scale.fit_transform(X_train)
```

```python
# build DNN model
import keras.models as k_models
from keras.layers.core import Dense, Activation
from keras.optimizers import Adam
from tensorflow import keras

model = k_models.Sequential()
model.add(Dense(input_dim = 15,units=40,activation='relu'))
model.add(Dense(units=100,activation='relu'))
model.add(Dense(units=5,activation='sigmoid'))
```

```python
# start train model
model.compile(loss="binary_crossentropy",optimizer='adam',metrics=['acc'])
model.fit(x=X_train,y=Y_train,batch_size=100,epochs=20)
```
